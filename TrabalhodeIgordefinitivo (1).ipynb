```python
import matplotlib.pyplot as plt
import numpy as np

# Dados para o Sudeste Asiático
x = [2017, 2018, 2019, 2020, 2021, 2022]
y1 = [2467263, 2602751, 1600599, 1494112, 1402635, 1322550]

# Dados para a América Latina e Caribe
y2 = [1753670, 1665465, 1619593, 1611732, 1548806, 1494199]

# Dados para a América do Norte
y3 = [189689, 188979, 188495, 189053, 188575, 189121]

# Largura das barras
bar_width = 0.25

# Posições das barras no eixo x
r1 = np.arange(len(x))
r2 = [i + bar_width for i in r1]
r3 = [i + bar_width for i in r2]

# Plotagem dos dados
plt.bar(r1, y1, color='orange', width=bar_width, edgecolor='grey', label='Sudeste Asiático')
plt.bar(r2, y2, color='blue', width=bar_width, edgecolor='grey', label='América Latina e Caribe')
plt.bar(r3, y3, color='red', width=bar_width, edgecolor='grey', label='América do Norte')

# Configurações do gráfico
plt.xlabel('Ano')
plt.ylabel('Número de Analfabetos (em milhões)')
plt.title('Analfabetismo entre jovens de 18-24 anos')
plt.xticks([r + bar_width for r in range(len(x))], x)
plt.legend()


plt.show()
```
